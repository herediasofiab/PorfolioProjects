#### 6.1 Tipo de usuarios por nivel de actividad <a class="anchor"  id="6.1"></a>


Determinaremos el tipo de usuarios con los datos que tenemos,los clasificaremos por actividad teniendo en cuenta la cantidad diaria de pasos. Los criterios estan basados en el siguiente artículo https://www.10000steps.org.au/articles/counting-steps/:

* Sedentarios - Menos de 5000 pasos al día.
* Poco activos - Entre 5000 y 7499 pasos al día.
* Activo - Entre 7500 y 9999 pasos al día.
* Muy activo - Más de 10000 pasos al día.


Para realizar la clasificación, primeramente agruparemos los datos diarios por el Id de usuario y calcularemos un promedio de los pasos realizados por persona, promedio de calorias y sueño, para un posterior análisis.

```r
promedio_diario <- actividad_sueño %>%
group_by(Id) %>%
summarise(media_pasos = mean(TotalSteps), media_calorias = mean(Calories), media_sueño = mean(TotalMinutesAsleep))

head(promedio_diario)
```

```r
tipo_de_usuario <- promedio_diario %>%
mutate (TipoUsuario = case_when(
    media_pasos < 5000 ~ "Sedentario",
    media_pasos  >= 5000 &  media_pasos <= 7499 ~ "Poco Activo",
    media_pasos >= 7500 & media_pasos <= 9999 ~ "Activo",
    media_pasos > 9999 ~ "Muy Activo") )

head(tipo_de_usuario)
```

* Ahora que tenemos una nueva columna con el tipo de usuario vamos a crear un marco de datos con el porcentaje de cada tipo de usuario para visualizarlos en un gráfico.

```r
porcentaje_usuario <- tipo_de_usuario %>%
group_by(TipoUsuario) %>%
summarise(n = n()) %>%
mutate(porcentaje = n/sum(n))%>%
mutate(porciento = scales::percent(porcentaje))


head(porcentaje_usuario)
```

```r
porcentaje_usuario %>%
  ggplot(aes(x="",y=porcentaje, fill=TipoUsuario)) +
  geom_bar(stat = "identity", linewidth = 3 )+
  coord_polar("y", start=2)+
  theme_minimal()+
  theme(legend.text = element_text (size = 20),
        legend.title = element_blank(),
        legend.key.size = unit(1.5, units = "cm"),
        axis.title.x= element_blank(),
        axis.title.y = element_blank(),
        panel.border = element_blank(), 
        panel.grid = element_blank(), 
        axis.ticks = element_blank(),
        axis.text.x = element_blank(),
        plot.title = element_text(hjust = 0.1, size=24, face = "bold")) +
  scale_fill_manual(values = c("#fd8f78","#E65CFF", "#7CD2D7", "#7369E0")) +
  geom_text(aes(label = porciento),
            position = position_stack(vjust = 0.5),color="white",size=9)+
  labs(title="Tipo de usuario según su actividad")
```

 * Podemos ver que ligeramente mayoria de los usuarios de nuestra muestra son activos.
 * Los usuarios se distribuyen equitativamente por su actividad teniendo en cuenta la cantidad diaria de pasos. 
 * Podemos determinar que, en función de la actividad de los usuarios, todos ellos utilizan dispositivos inteligentes.

#### 6.2 Pasos contados por día de la semana <a class="anchor"  id="6.2"></a>

Ahora vamos a determinar qué días de la semana los usuarios son más activos.
También vamos a comprobar si los usuarios caminan la cantidad recomendada de 7500 pasos por dia.
A continuación calculamos los días de la semana en función de la fecha. También estamos calculando el promedio de pasos caminados.

```r
dias_actividad_sueño <- actividad_sueño %>%
mutate(dias = weekdays (Fecha))

dias_actividad_sueño$dias <-ordered(dias_actividad_sueño$dias, levels=c("Monday", "Tuesday", "Wednesday", "Thursday",
"Friday", "Saturday", "Sunday"))
```

```r

diastabla1 <- diastabla %>%
  group_by(DiasSemana) %>%
  summarize(PasosTotales = mean(TotalSteps), SueñoTotal = mean(TotalMinutesAsleep))

head(diastabla1)
```

#### 6.4 Pasos por hora <a class="anchor"  id="6.4"></a>


Para mayor profundidad de análisis, vamos a graficar la cantidad de pasos efectuados a lo largo dia, para identificar los horarios con mayor actividad de los usuarios. Para ello utilizaremos la tabla "pasos" en la cual encontramos los pasos por fecha, que dividiremos en fecha y otra columna hora.

```r
pasos1 <- pasos %>%
separate(Fecha, into=c("fecha", "hora"),sep =" " ) %>%
group_by(hora) %>%
summarize(Pasos_Promedio= mean(StepTotal))

pasos2 <- pasos1 %>%
separate(hora, into=c("hora", "minuto"),sep =":00:00") %>%
select(hora,Pasos_Promedio)
head(pasos2)

```

```r
pasos2 %>%
ggplot() +
  geom_col(mapping = aes(x=hora, y =Pasos_Promedio, fill =Pasos_Promedio)) + 
  labs(title = "Pasos por hora a lo largo del dia", x="", y="") + 
  scale_fill_gradient(low ="#39D2D7", high ="#E65CFF")+
  theme(axis.text.x = element_text(angle = 90),
        legend.text = element_text (size = 18),
        legend.title = element_text(size = 15, face = 1),
        axis.text = element_text(size = 15, face = "bold"),
        legend.key.size = unit(1.5, units = "cm"),
        panel.border = element_rect(colour = "black", fill=NA),
        plot.title = element_text(hjust = 0.1, size=23, face = "bold"))
```

Observando el gráfico podemos determinar lo siguiente:
* Los usuarios realizan la mayor cantidad de pasos promedio por hora a las 18:00.
* Se observa que los usuarios son más activos entre las 8 hs y las 19hs. 

#### 6.5 Correlación sueño y pasos diarios <a class="anchor"  id="6.5"></a> 

* Ahora determinaremos el nivel de correlación entre el tiempo de sueño y pasos diarios, para determinar si la actividad fisica de nuestros usuarios mejora la cantidad de sueño y viceversa.

```r
ggplot(data=actividad_sueño)+
 geom_smooth(mapping=aes(x=TotalSteps, y=TotalMinutesAsleep),color ="#00CDAF")+
 geom_point(mapping=aes(x=TotalSteps, y=TotalMinutesAsleep, alpha=TotalMinutesAsleep),color="purple", size=5)+
 labs(title = "Pasos vs Sueño", x = "Pasos Diarios", y= "Minutos de Sueño") +
   theme(panel.background = element_blank(),
        plot.title = element_text(size=25, hjust=0.5),
        axis.text = element_text(size = 10, face = "bold"),
        axis.title.x = element_text(family='Anton', hjust=0.5, size=18),
        axis.title.y = element_text(family='Anton', hjust=0.5, size=18),
        panel.border = element_rect(colour = "black", fill=NA),
        legend.position = "none",)
```

```r
cor.test(actividad_sueño$TotalSteps,actividad_sueño$TotalMinutesAsleep, method = 'pearson', conf.level = 0.95)
```

 * El coheficiente de correlacion es menor a 0, mostrando en el grafico una ligera pendiente negativa. Con ello, podemos afirmar que existe una  pequeña relación negativa entre la cantidad de pasos y minutos de sueño diarios.
 * Con un nivel de confianza del 95%, la correlación entre las variables es de -0.19 Esto significa que existe una débil relación entre las variables.
 
* **No existe correlación suficiente entre el nivel de actividad diaria basado en los pasos y la cantidad de minutos que los usuarios duermen al día.**



#### 6.6 Correlación calorías y pasos diarios <a class="anchor"  id="6.6"></a>

#### 6.7 Correlación entre minutos que tardan los usuarios en conciliar el sueño vs pasos diarios <a class="anchor"  id="6.7"></a> 

A continuación evaluaremos la correlación entre los pasos realizados por los usuarios en un dia y el tiempo que les lleva conciliar el sueño. Para ello restaremos las columnas, minutos totales en cama, menos el minutos totales dormidos, a este resultado lo correlacionaremos a la cantidad de pasos diarios.

```r
pasos_sueño <- actividad_sueño %>%
mutate(minutos_dormirse = TotalTimeInBed - TotalMinutesAsleep)%>%
select(Id,minutos_dormirse,TotalSteps,TotalTimeInBed)

head(pasos_sueño)
summarise(pasos_sueño,prom_min_dormirse = mean(minutos_dormirse))
```

```r
ggplot(data=pasos_sueño)+
 geom_point(mapping=aes(x=TotalSteps, y=minutos_dormirse, alpha=TotalSteps),color="#B7106C", size=5)+
 labs(title = "Pasos diarios y Minutos tardan en dormirse", x = "Pasos Diarios", y= "Minutos tardan en dormir") +
 scale_y_continuous(breaks = seq(0,500, by = 50))+
 scale_x_continuous(breaks = seq(200,50000, by = 2500))+
 geom_hline(yintercept = 50) +
   theme(plot.title = element_text(size=23, hjust=0.5),
        axis.text = element_text(size = 10, face = "bold"),
        axis.title.x = element_text(family='Anton', hjust=0.5, size=18),
        axis.title.y = element_text(family='Anton', hjust=0.5, size=18),
        panel.border = element_rect(colour = "black", fill=NA),
        legend.position = "none",)
```

#### 6.8 Días de uso del dispositivo inteligente <a class="anchor"  id="6.8"></a> 

Vamos a analizar cuál es la frecuencia de uso de los dispositivo inteligentes, de los usuarios de nuestra muestra. 
Vamos a calcular el número de usuarios que utilizaron su dispositivo inteligente a diario, clasificando nuestra muestra en tres categorías sabiendo que el máximo es 31 días:

* Alto: Usuarios que utilizaron su dispositivo entre 21 y 31 días.
* Medio: usuarios que utilizaron su dispositivo entre 10 y 20 días.
* Bajo: usuarios que utilizaron su dispositivo entre 1 y 10 días.

Para ello crearemos un nuevo marco de datos agrupando por Id, calculando el número de días de uso y su respectiva clasificación.

```r
diasdeuso <- actividad_sueño %>%
group_by(Id) %>%
summarise(cantidad_dias = sum(n())) %>%
mutate (uso_diario = case_when(
    cantidad_dias >= 21 & cantidad_dias <= 31 ~ "Alto",
    cantidad_dias >= 11 & cantidad_dias <= 20 ~ "Medio",
    cantidad_dias >= 1 & cantidad_dias <= 10 ~ "Bajo") )

head(diasdeuso) 
```

```r
diasdeuso_porcen <- diasdeuso %>%
group_by(uso_diario) %>%
summarise(n = n()) %>%
mutate(porcentaje = n/sum(n))%>%
mutate(porciento = scales::percent(porcentaje))


head(diasdeuso_porcen)
```

```r
diasdeuso_porcen %>%
  ggplot(aes(x="",y=porciento, fill=uso_diario)) +
  geom_bar(stat = "identity", width = 20, size = 10,)+
  coord_polar("y", start=3.13)+
  theme_minimal()+
  theme(legend.text = element_text (size = 18),
        legend.title = element_blank(),
        legend.key.size = unit(1, units = "cm"),
        axis.title.x= element_blank(),
        axis.title.y = element_blank(),
        panel.border = element_blank(), 
        panel.grid = element_blank(), 
        axis.ticks = element_blank(),
        axis.text.x = element_blank(),
        plot.title = element_text(hjust = 0, size=24, face = "bold")) +
  scale_fill_manual(values = c("#fd8f78","#7CD2D7", "#B7106C"), 
                    labels = c("Alto (21-31 días)",
                               "Medio (11-20 días)",
                               "Bajo (1-10 días)")) +
  geom_text(aes(label = porciento),
            position = position_stack(vjust = 0.5),color="white",size=9)+
  labs(title="Uso diario del dispositivo inteligente")
```

Analizando nuestros resultados podemos ver que:

* El 50% de los usuarios de nuestra muestra utiliza su dispositivo con frecuencia: entre 21 y 31 días.
* El 12% utiliza su dispositivo entre 11 y 20 días.
* El 38% de nuestra muestra utiliza su dispositivo con muy poca frecuencia.

#### 6.9 Horas de uso del dispositivo inteligente <a class="anchor"  id="6.9"></a> 

Ya hemos investigado la cantidad de días que usan el dispositivo inteligente los usuarios, ahora queremos determinar cuantas horas usan el dispositivo los usuarios, para ello clasificaremos en 3 categorías:

* Todo el dia 
* Más de medio dia 
* Menos de medio dia 

```r
fusion_horas <- merge(actividad, diasdeuso, by=c ("Id"))
```

```r
horas_uso <- fusion_horas %>%
mutate(minutos_uso =(VeryActiveMinutes + FairlyActiveMinutes + LightlyActiveMinutes + SedentaryMinutes))%>%
mutate(porcen_uso= (minutos_uso/1440)*100) %>%
mutate(etiqueta = case_when(
    porcen_uso == 100 ~ "Todo el dia",
    porcen_uso >= 50 & porcen_uso < 100 ~ "Más de medio dia",
    porcen_uso < 50 & porcen_uso > 0 ~ "Menos de medio dia")) %>%
select(Id, minutos_uso,porcen_uso, etiqueta)


head(horas_uso)
```

```r
etiquetas_hora <- horas_uso %>%
group_by(etiqueta) %>%
summarise(n = n()) %>%
mutate(porcentaje = n/sum(n))%>%
mutate(porciento = scales::percent(porcentaje))

head(etiquetas_hora) 
```

```r
etiquetas_hora %>%
  ggplot(aes(x="",y=porcentaje, fill=etiqueta)) +
  geom_bar(stat = "identity", width = 20, size = 10,)+
  coord_polar("y", start=3.13)+
  theme_minimal()+
  theme(legend.text = element_text (size = 18),
        legend.title = element_blank(),
        legend.key.size = unit(1, units = "cm"),
        axis.title.x= element_blank(),
        axis.title.y = element_blank(),
        panel.border = element_blank(), 
        panel.grid = element_blank(), 
        axis.ticks = element_blank(),
        axis.text.x = element_blank(),
        plot.title = element_text(hjust = 0, size=24, face = "bold")) +
  scale_fill_manual(values = c("#7CD2D7","#7369E0","#FC5677"), 
                    labels = c("Más del medio dia","Menos del medio día","Todo el día")) +
  geom_text(aes(label = porciento),
            position = position_stack(vjust = 0.4),color="white",size=9)+
  labs(title="Uso diario del dispositivo inteligente")
```

Luego de realizar el anterior análisis podemos afirmar :
* El 60% de los usuarios lleva puesto el dispositivo más del medio día.
* El 36% de los usuarios lleva puesto el dispositivo todo el día.
* El  4% de los usuarios lleva puesto el dispositivo menos del medio día.
