## 1. Summary <a class="anchor"  id="1."></a> 

Stack Overflow, founded in 2008 by Jeff Atwood and Joel Spolsky, is a leading question-and-answer site aimed at both professional and amateur programmers. As part of its annual practice, the platform conducts a survey of its users, addressing a variety of programming-related topics. This project has the goal of analysing the survey responses to understand the evolution of the developer experience, identify the most popular technologies, and examine the employment situation in the field.


## 2. Preparation <a class="anchor"  id="2."></a> 

#### 2.1 Data storage  <a class="anchor"  id="2.1"></a> 

The data source used can be found on the official Stack Overflow website, at the following link: https://insights.stackoverflow.com/survey 


#### 2.2 Accessibility and privacy of data  <a class="anchor"  id="2.2"></a> 

The database is under an open source licence.(ODbL): http://opendatacommons.org/licenses/odbl/1.0/. 
This licence lets users share, modify and use it freely.

#### 2.3 Data set information  <a class="anchor"  id="2.3"></a> 
The dataset comprises more than 64,000 responses to a set of 60 questions, although it should be noted that many of these questions are optional. To preserve privacy, responses have been anonymised, removing any personally identifiable information. Each respondent has been assigned a random ID in order to maintain the confidentiality of the participants.

 #### 2.4 Data organisation and verification  <a class="anchor"  id="2.4"></a> 
 The dataset consists of 2 CSV documents:
   1. survey_results_public.csv - CSV file with the main results of the survey, one respondent per row and one column per answer.
    
   2. survey_results_schema.csv - CSV file with the outline of the survey, i.e. the questions that correspond to each column name.
    
#### 2.5 Credibility and integrity of data <a class="anchor"  id="2.5"></a> 
In May 2020, more than 90,000 developers participated in the annual survey about their work environment and the tools they use, as well as their preferences. Participants were primarily recruited through various channels associated with Stack Overflow. The five principal sources of recruitment included on-site posts, blog posts, email lists, meta.stackoverflow posts, banner ads, and social media posts. Given that respondents were recruited in this way, it is possible that users highly engaged with Stack Overflow were more likely to notice the links to the survey and click to participate. Also, given that the survey was only available in English, there is the possibility of a slight sampling bias in terms of the overall representation of programmers.

## 4. Data Cleaning <a class="anchor"  id="4."></a>

* Demographic data from surveys and the global developer community.
* Distribution of skills, experience and programming preferences.
* Employment-related information, preferences and opinions.

Let's select a subset of columns with the relevant data for our analysis.

```python
columns = [
    # Demographics
    'Country',
    'Age',
    'Gender',
    'EdLevel',
    'UndergradMajor',
    # Programming experience
    'Hobbyist',
    'Age1stCode',
    'YearsCode',
    'YearsCodePro',
    'LanguageWorkedWith',
    'LanguageDesireNextYear',
    # Work
    'Employment',
    'DevType',
    'WorkWeekHrs',
    'JobSat',
    'JobFactors',
    'NEWOvertime',
    'NEWEdImpt',
    'ConvertedComp'
]
```

### 4.2 Data frame copy <a class="anchor"  id="4.2"></a> 
We will copy the information contained in these columns to a new dataset called survey_df. In this way, we can make further modifications without altering the original data structure.

```python
survey_df = survey_df[columns].copy()
```

### 4.3 Modification of data format  <a class="anchor"  id="4.3"></a>

```python
survey_df.info()
```

Most of the columns have 'object' data types except for three columns (WorkWeekHrs, ConvertedComp and Age) whose data type is numeric 'float64'. The columns (Age1stCode, YearsCode, YearsCodePro) are under 'object' format but contain numeric data, we will correct them below.

```python
survey_df['Age1stCode'] = pd.to_numeric(survey_df.Age1stCode, errors='coerce')
survey_df['YearsCode'] = pd.to_numeric(survey_df.YearsCode, errors='coerce')
survey_df['YearsCodePro'] = pd.to_numeric(survey_df.YearsCodePro, errors='coerce')
```

### 4.4 Correction of age range and hours per week  <a class="anchor"  id="4.4"></a>

```python
survey_df.describe()
```

We can notice in the column (Age) that the maximum age is (279 and 1) the minimum, this is a common typing error in surveys, we will correct it by filtering as maximum 100 years and minimum age 10. The column (WorkWeekHrs) will be filtered to a maximum of 140 hours per week.

```python
survey_df.drop(survey_df[survey_df.Age< 10].index, inplace=True)
survey_df.drop(survey_df[survey_df.Age> 100].index, inplace=True)
survey_df.drop(survey_df[survey_df.WorkWeekHrs>140].index, inplace=True)
```

### 6.1 What are the most popular programming languages ? <a class="anchor" id="6.1"></a>

### 6.2 What are the most popular programming languages for Data Science workers? <a class="anchor" id="6.2"></a>

We are going to identify which are the most used programming languages for data analytics, data science or business analytics. To do so, we will use the new data frame "work_programs" created in the previous point, but this time we will filter with the column "DevType", remembering that it contains the answers of the respondents' jobs, establishing as a restriction that only the most popular languages among data experts appear.

```python
languages_job = languages.assign(DevType=languages.DevType.str.split(";")).explode('DevType')
```

```python
filter = ['Data or business analyst', 'Data scientist or machine learning specialist']
data_languages = languages_job[languages_job['DevType'].isin(filter)]
data_languages.head(3)
```

```python
count_data_languages = data_languages.LanguageWorkedWith.value_counts().sort_values(ascending=False)
```

```python
# Graph most popular data languages for data workers
from matplotlib.colors import LinearSegmentedColormap
colors = ['slateblue','darkgrey']  
n_bins = 31
cmap = LinearSegmentedColormap.from_list('mi_degradado', colors, N=n_bins)

total = count_data_languages.sum()
percentages = (count_data_languages / total) * 100

plt.figure(figsize=(10, 8))
bars = plt.barh(list(percentages.index),count_data_languages, color=cmap(np.linspace(0, 1, len(count_data_languages))))

plt.xlabel('Number of workers data')
plt.ylabel('Languages')
plt.title('What are the most popular programming languages for Data Science workers?')

for i, count in enumerate(count_data_languages):
    percent = percentages.loc[count_data_languages.index[i]]
    plt.text(count + 100, i, f'{percent:.2f}%', va='center')

plt.show()
```

* As a result of the analysis we can see that the two most popular languages among data workers are almost tied with SQL due to its great capacity to handle large databases and Python, which has grown in popularity due to the large number of uses both for creating graphs, ease of learning and number of libraries that facilitate the management of data frames.

### 6.8 Job Satisfaction vs Salary  <a class="anchor" id="6.8"></a>
As we have already seen in the previous analysis, programmers are positioned with good salaries and most of them are satisfied with their job. We will identify what is the level of relationship between job satisfaction and salary in developers. To do this we will rank each level of satisfaction from (1 to 5) with 5 being the highest level of job satisfaction.

```python
job_position['point_sat'] = job_position['JobSat'].apply(lambda texto :
    5 if (texto == 'Very satisfied') else
    4 if (texto == 'Slightly satisfied') else
    3 if (texto == 'Neither satisfied nor dissatisfied') else
    2 if (texto == 'Slightly dissatisfied') else
    1 if (texto == 'Very dissatisfied') else None)
```

```python
#Graph Job Satisfaction vs Salary
plt.figure(figsize=(15, 5))
plt.title('Job Satisfaction vs Salary')

colors = sns.blend_palette(['slateblue', 'darkturquoise'], as_cmap=False, n_colors=5)
scatter = sns.scatterplot(x='point_sat', y='ConvertedComp', hue='point_sat', s=80, data=job_position, palette=colors, legend=False)

sns.regplot(x='point_sat', y='ConvertedComp', data=job_position, scatter=False, color='mediumvioletred')

plt.show()
```

```python
correlation_spearman = job_position['point_sat'].corr(job_position['ConvertedComp'], method='spearman')
correlation_spearman 
```

* As we can see in the graph, the distribution of responses in terms of level of satisfaction is equal from the lowest to the highest salary. We also performed a correlation coefficient, using the spearman method, which indicates how closely related two variables are.

* The correlation coefficient is greater than 0, showing a slight positive slope in the graph. Thus, we can say that there is a positive relationship between salary and level of satisfaction.
The correlation between the variables is (0.14). This means that the relationship between the variables is minimal. 

* **There is not enough correlation between the variables to affirm that with an increase in salary, job satisfaction will increase. With this we can conclude that the high percentage of job satisfaction in the field of programmers is not due to salaries as a determining variable.**

```python
#Graphic important aspects for accepting a job
total = count_aspects.sum()
percent = (count_aspects / total) * 100

colors = sns.blend_palette(['slateblue','darkturquoise'], as_cmap=False, n_colors=11)

plt.figure(figsize=(13, 12))
sns.barplot(x=count_aspects, y=count_aspects.index, order=count_aspects.index, palette=colors)
plt.xlabel('Number of Respondens')
plt.ylabel('Labor Aspects')
plt.title('What are the most important aspects of a job apart from salary, benefits and location?')

for i, count in enumerate(count_aspects):
    percentage= percent.loc[count_aspects.index[i]]
    plt.text(count + 100, i, f'{percentage:.2f}%', va='center')

plt.show()
```
