## 1. Resumen <a class="anchor"  id="1."></a> 

Stack Overflow, fundado en 2008 por Jeff Atwood y Joel Spolsky, es un destacado sitio de preguntas y respuestas dirigido a programadores tanto profesionales como aficionados. Como parte de su práctica anual, la plataforma realiza una encuesta a sus usuarios, abordando una variedad de temas relacionados con la programación. Este proyecto tiene como objetivo analizar las respuestas de la encuesta para entender la evolución de la experiencia de los desarrolladores, identificar las tecnologías más populares y examinar la situación laboral en el campo.

#### 2.3 Información sobre el conjunto de datos   <a class="anchor"  id="2.3"></a> 
El dataset comprende más de 64,000 respuestas a un conjunto de 60 preguntas, aunque cabe destacar que muchas de estas preguntas son opcionales.  El conjunto de datos se encuentra formado por 2 documentos CSV:

#### 2.5 Credibilidad e integridad de los datos  <a class="anchor"  id="2.5"></a> 
En mayo de 2020, más de 90,000 desarrolladores participaron en la encuesta anual sobre su entorno laboral y las herramientas que utilizan, así como sus preferencias. Los participantes fueron principalmente reclutados a través de diversos canales asociados con Stack Overflow. Las cinco fuentes principales de reclutamiento incluyeron mensajes dentro del sitio, publicaciones en blogs, listas de correo electrónico, publicaciones en meta.stackoverflow, anuncios en banners y publicaciones en redes sociales. Dado que los encuestados fueron reclutados de esta manera, es posible que los usuarios altamente comprometidos con Stack Overflow fueran más propensos a notar los enlaces para la encuesta y hacer clic para participar. Además, teniendo en cuenta que la encuesta solo estaba disponible en inglés, existe la posibilidad de un sesgo de muestreo leve en términos de la representación global de programadores.

## 4. Limpieza de Datos <a class="anchor"  id="4."></a>

Si bien las respuestas de la encuesta contienen una gran cantidad de información, limitaremos nuestro análisis a las siguientes áreas:

* Datos demográficos de los encuestados y de la comunidad de programación global
* Distribución de habilidades, experiencia y preferencias de programación.
* Información, preferencias y opiniones relacionadas con el empleo.

Seleccionemos un subconjunto de columnas con los datos relevantes para nuestro análisis.

```python
columnas = [
    # Demográficas
    'Country',
    'Age',
    'Gender',
    'EdLevel',
    'UndergradMajor',
    # Experiencia en programación
    'Hobbyist',
    'Age1stCode',
    'YearsCode',
    'YearsCodePro',
    'LanguageWorkedWith',
    'LanguageDesireNextYear',
    # Trabajo
    'Employment',
    'DevType',
    'WorkWeekHrs',
    'JobSat',
    'JobFactors',
    'NEWOvertime',
    'NEWEdImpt',
    'ConvertedComp'
]
```

### 4.2 Copia del data frame <a class="anchor"  id="4.2"></a> 
Vamos a copiar la información contenida en estas columnas a un nuevo conjunto de datos llamado encuesta_df. De esta manera, podemos realizar más modificaciones sin alterar la estructura original de datos.

```python
encuesta_df = encuesta_df[columnas].copy()
```

### 4.3 Modificación del formato de datos <a class="anchor"  id="4.3"></a>

```python
encuesta_df.info()
```

La mayoria de las columnas tienen tipos de datos 'object' excepto tres columnas (WorkWeekHrs, ConvertedComp y Age) las cuales su tipo de datos es numérico 'float64'. Las columnas (Age1stCode, YearsCode, YearsCodePro) están bajo formato 'object' pero contienen datos numéricos, a continuación vamos a correguirlas.

```python
encuesta_df['Age1stCode'] = pd.to_numeric(encuesta_df.Age1stCode, errors='coerce')
encuesta_df['YearsCode'] = pd.to_numeric(encuesta_df.YearsCode, errors='coerce')
encuesta_df['YearsCodePro'] = pd.to_numeric(encuesta_df.YearsCodePro, errors='coerce')
```

### 4.4 Corrección del rango de edad y horas por semana <a class="anchor"  id="4.4"></a>

```python
encuesta_df.describe()
```

Podemos notar en la columna (Age) que la edad máxima es (279 y 1) la mínima, este es un error que puede ser de tipeo común en las encuestas, lo vamos a correguir filtrando como máximo 100 años y edad minima 10. La columna (WorkWeekHrs) la filtraremos como máximo a 140 hs por semana.


```python
encuesta_df.drop(encuesta_df[encuesta_df.Age< 10].index, inplace=True)
encuesta_df.drop(encuesta_df[encuesta_df.Age> 100].index, inplace=True)
encuesta_df.drop(encuesta_df[encuesta_df.WorkWeekHrs>140].index, inplace=True)
```

### 5.5 Nivel de Educación <a class="anchor" id="5.5"></a>
La formación académica en el campo de la informática se suele ver como un requisito fundamental para ingresar al mundo de la programación. No obstante, existen numerosos recursos y tutoriales gratuitos en línea que permiten adquirir habilidades de programación sin necesidad de educación formal. Podemos analizar los niveles educativos de las personas encuestadas para obtener una perspectiva más amplia sobre este tema.

```python
# Gráfico Nivel de Educación
plt.figure(figsize=(12,8))
conteo_edu = encuesta_df['EdLevel'].value_counts().sort_values(ascending=False) # Agrupamos y contamos según su nivel educativo
porcen = conteo_edu.index
total = conteo_edu.values.sum()
colors = sns.blend_palette(['cyan', 'slateblue'], as_cmap=False, n_colors=9)

sns.countplot(data=encuesta_df, y='EdLevel', order=porcen, palette=colors)
plt.xlabel('Número de Encuestados')
plt.title('Nivel Educativo de los Encuestados')
plt.ylabel(None)

for i, contador in enumerate(conteo_edu):
   pcnt = '{:0.1f}%'.format(100*contador/total)
   plt.text(contador-20, i, pcnt, va ='center')
```

Podemos notar que la mayoria de los programadores encuestados poseen una licenciatura o un master. Si sumamos los encuestados que tienen una licenciatura (46.6%) más los que tienen un master (22.4%) nos da un total de (69,25%), es decir que la mayoria de los encuestados poseen educación universitaria completa.

### 5.9 Puestos de trabajo de los programadores <a class="anchor" id="5.9"></a>
En el próximo paso, exploraremos los roles ocupados por los encuestados. Para llevar a cabo esta tarea, nos valdremos de la columna "DevType", que proporciona detalles sobre las funciones desempeñadas por los participantes. Dado que la pregunta admite respuestas múltiples, la columna contiene listas de valores separados por punto y coma (;), lo cual complica el análisis directo. Por esta razón, procederemos a generar una tabla que refleje los distintos puestos de trabajo.

```python
preguntas.DevType
```

```python
encuesta_df.DevType
```

```python
trabajo = encuesta_df.assign(DevType=encuesta_df.DevType.str.split(";")).explode('DevType')
trabajo.head(3)
```

```python
conteo_trabajo = trabajo.DevType.value_counts().sort_values(ascending=False)
conteo_trabajo
```

```python
#Gráfico puestos de trabajo de los programadores
from matplotlib.colors import LinearSegmentedColormap
colors = ['slateblue','mediumvioletred']  
n_bins = 31
cmap = LinearSegmentedColormap.from_list('mi_degradado', colors, N=n_bins)

plt.figure(figsize=(10, 8))

total = conteo_trabajo.sum()
porcentajes = (conteo_trabajo / total) * 100

bars = plt.barh(list(porcentajes.index), conteo_trabajo, color=cmap(np.linspace(0, 1, len(conteo_trabajo))))

plt.xlabel('Cantidad de Programadores')
plt.ylabel('Profesion')
plt.title('Puestos de trabajo de los programadores')

for i, cantidad in enumerate(conteo_trabajo):
    porcentaje = porcentajes.iloc[i]
    plt.text(cantidad + 100, i, f'{porcentaje:.2f}%', va='center')

plt.show()
```

### 6.8 Satisfacción laboral vs Salario <a class="anchor" id="6.8"></a>
Como ya hemos visto en el análisis anterior, los programadores se encuentran posicionados con buenos salarios y la mayoría se encuentra satisfecha con su trabajo.Identificaremos cuál es el nivel de relación entre satisfacción laboral y salario en los desarrolladores. Para ello clasificaremos cada nivel de satisfacción del (1 al 5) siendo 5 el nivel mayor satisfacción laboral.

```python
trabajo['puntos_sat'] = trabajo['JobSat'].apply(lambda texto :
    5 if (texto == 'Very satisfied') else
    4 if (texto == 'Slightly satisfied') else
    3 if (texto == 'Neither satisfied nor dissatisfied') else
    2 if (texto == 'Slightly dissatisfied') else
    1 if (texto == 'Very dissatisfied') else None)
```

```python
#Gráfico Satisfacción Laboral vs Salario
plt.figure(figsize=(15, 5))
plt.title('Satisfacción Laboral vs Salario')

colors = sns.blend_palette(['slateblue', 'darkturquoise'], as_cmap=False, n_colors=5)
scatter = sns.scatterplot(x='puntos_sat', y='ConvertedComp', hue='puntos_sat', s=80, data=trabajo, palette=colors, legend=False)

sns.regplot(x='puntos_sat', y='ConvertedComp', data=trabajo, scatter=False, color='mediumvioletred')

plt.show()
```

```python
correlation_spearman = trabajo['puntos_sat'].corr(trabajo['ConvertedComp'], method='spearman')
correlation_spearman 
```

Como podemos ver en el gráfico, la distribución de las respuestas en cuanto a nivel de satisfación es equitativa desde el salario más bajo al más alto. También realizamos un coheficiente de correlación, con el método spearman, que indica cuán relacionadas estan dos variables.

El coheficiente de correlación es mayor a 0, mostrando en el gráfico una ligera pendiente positiva. Con ello, podemos decir que existe una relación positiva entre el salario y el nivel de satisfacción.
La correlación entre las variables es de (0.14) Esto significa que relación entre las variables es mínima. 

No existe correlación suficiente entre las variables para afirmar que ante un aumento del salario, aumentará la satisfacción laboral. Con esto podemos concluir que el gran porcentaje de satisfacción laboral en el ámbito de los programadores no esta dado por los salarios, como variable determinante.
